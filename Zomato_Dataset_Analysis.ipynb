```python
df_country = pd.read_excel('dataset/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df['Country'].value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
# top 3 countries uses zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%');
```

Observation: Major market of zomato is in India. 
After that USA and then United Kingdoms.

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'})
ratings
```

## Observation
1. When Rating is between 4.5 to 4.9 ---> Excellent
2. When Rating are between 4.0 to 4.4 ---> Very Good
3. When Rating is between 3.5 to 3.9 ---> Good
4. When Rating is between 2.5 to 3.4 ---> Average
5. When Rating is between 1.8 to 2.4 ---> Poor
6. When Rating is 0.0 ---> Not Rated

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color", data=ratings, palette=['wheat','red','orange','yellow', 'green','green']);
```

Observation:
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4
